```python
import pandas as pd
import numpy as np
import unicodedata
```

```python
directory_path = "D:\\#Data\\Scraped Data\\Adventus\\Final"
filename = "adventus_india_data.csv"

df = pd.read_csv(f"{directory_path}\\{filename}", encoding="utf-8")
df.head()
```

```python
def replace_special_characters(text):
    return unicodedata.normalize('NFKD', text).encode('ASCII', 'ignore').decode('ASCII')
```

```python
df['state'] = df['state'].apply(replace_special_characters)
df.to_csv(f"{directory_path}\\{filename}", index=False, encoding='utf-8')
df.head()
```
