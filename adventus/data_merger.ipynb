```python
import numpy as np
import pandas as pd
import os
```

```python
directory_path = "D:\\#Data\\Scraped Data\\Adventus"
data_frames = []
```

```python
for file in os.listdir(directory_path):
    if file.endswith(".csv"):
        file_path = os.path.join(directory_path, file)
        df = pd.read_csv(file_path)
        data_frames.append(df)
```

```python
merged_df = pd.concat(data_frames, ignore_index=True)
```

```python
len(merged_df["state"].unique())
```

```python
state_counts = merged_df.groupby("state").size()
state_counts
```

```python
adventus_india_data = merged_df.groupby('course_id').agg({
    'course_name': 'first',
    'course_url': 'first',
    'college': 'first',
    'location': 'first',
    'duration': 'first',
    'turnaround_time': 'first',
    'fees': 'first',
    'intakes': 'first',
    'state': lambda x: ', '.join(set(x))
}).reset_index()
```

```python
adventus_india_data.head()
```

```python
output_file_path = 'D:\\#Data\\Scraped Data\\Adventus\\Final\\adventus_india_data.csv'
adventus_india_data.to_csv(output_file_path, index=False)
```
